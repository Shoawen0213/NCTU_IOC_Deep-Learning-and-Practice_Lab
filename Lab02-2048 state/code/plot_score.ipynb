```python
def get_per1000k(dataframe):
    data_all0 = np.zeros(int(len(dataframe)/1000))
    data_per1K_0 = pd.DataFrame(data_all0)
    print(len(data_per1K_0))
    k=0
    for i in range(len(dataframe)):
        if(i!=0) and (i%1000==0):
#             print(i)
#             print(k)
            data_per1K_0[0][k] = int(dataframe[0][i])
            k += 1
    print(f'length of revise:{len(data_per1K_0)}')
    print("Cleaning!!!!")
    data_per1K_clr = data_per1K_0[0][0:]
    data_per1K_clr
    return data_per1K_clr
```
